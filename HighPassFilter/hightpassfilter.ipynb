```python
import numpy as np
import math
import matplotlib.pyplot as plt

f=1000 #equivalent to 1KHz
R=430 #430 ohms
C=1e-6 #1microFarad
v=5  #5 v
t=np.linspace(0,0.002,1000)
w=2*np.pi*f
theta=np.pi/2-np.arctan(w*R*C)
v2_mag=((w*R*C)/np.sqrt(1+(w*R*C)**2))*v
v1=v*np.cos(w*t)
v2=v2_mag*np.cos(w*t+theta)
plt.plot(t,v1,label='input signal')
plt.plot(t,v2,label='output signal')
plt.title("Plotting the input and output signal")
plt.legend()
plt.grid()
plt.show()
print(f"The phase in degrees: {math.degrees(theta)} {theta} ")
```

plotting magnitude against angular frequency

```python
w=np.linspace(0,10000,1000)
v2_mag=(w*R*C/np.sqrt(1+(w*R*C)**2))*v
cut_off_w=(1/(R*C))
cut_off_mag=(cut_off_w*R*C)/np.sqrt(1+(cut_off_w*R*C)**2)*v
plt.plot(w,v2_mag)
plt.title("Magnitued vs w")
plt.xlabel("w")
plt.ylabel("Magnitude")
#Drawing a dotted line (value of the cutoff frequency)
plt.axvline(x=cut_off_w, color='r', linestyle='--', label='Cutoff Frequency')
plt.axhline(y=cut_off_mag, color='g', linestyle='--', label='Cutoff Magnitude')
plt.grid()
plt.show()
print(f"Magnitude at Half power Frequency:{cut_off_mag}")
```
